##**1 - Lendo arquivos CSV**

```python
import pandas as pd
```

Foi necessário realizar a verificação do arquivo pois ele foi salvo na codificação 'ISO-8859-1' e suas 3 primeiras linhas eram compostas por metadados o que poderia causar um erro na leitura do arquivo.

```python
file_path = '/content/dados_sus.csv'

try:
    with open(file_path, 'r', encoding='ISO-8859-1') as f:
        for i, line in enumerate(f):
            print(f"Line {i+1}: {line.strip()}")
            if i > 10:
                break
except FileNotFoundError:
    print(f"Error: File not found at {file_path}")
except Exception as e:
    print(f"An error occurred: {e}")
```

###**Parâmetros da função read_csv**

```python
dados = pd.read_csv('/content/dados_sus.csv', encoding='ISO-8859-1', sep=';', skiprows=3, header=0)
display(dados.head())
```

```python
dados
```

```python
indices_para_deletar = [28, 29, 30, 31, 32, 33, 34]
dados = dados.drop(indices_para_deletar)

display(dados.tail())
```

```python
dados
```

###**Escrevendo arquivos CSV**

```python
dados.to_csv('dados_sus_tratados.csv', index=False)
```

##**5 - Trabalhando com banco de dados**

###**Criando um banco de dados local**

```python
import sqlalchemy
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
engine = create_engine('sqlite:///:memory:')
```

###**Escrevendo em um banco de dados**

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
```

```python
dados_banco = pd.read_csv(url)
dados_banco
```

```python
dados_banco.head()
```

```python
dados_banco.to_sql('clientes', engine, index=False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

```python
query = 'SELECT * FROM clientes'
```

```python
pd.read_sql(query, engine)
```

```python
from sqlalchemy import text
```

```python
query = 'DELETE FROM clientes WHERE ID_Cliente=5008809'
with engine.connect() as conn:
    result = conn.execute(text(query))
    conn.commit()
```

```python
pd.read_sql_table('clientes', engine)
```

```python
query = 'UPDATE clientes SET Rendimento_anual="300000" WHERE ID_Cliente=6840104'
with engine.connect() as conn:
    result = conn.execute(text(query))
    conn.commit()
```

```python
pd.read_sql_table('clientes', engine).tail()
```
